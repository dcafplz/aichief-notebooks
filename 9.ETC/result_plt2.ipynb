### 초해상화 전후 결과 비교

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def imwrite(filename, img, params=None):
    try:
        ext = os.path.splitext(filename)[1]
        result, n = cv2.imencode(ext, img, params)

        if result:
            with open(filename, mode='w+b') as f:
                n.tofile(f)
            return True
        else:
            return False
    except Exception as e:
        print(e)
        return False
```

```python
path_train = 'crop'
for i in os.listdir(path_train):
    src = np.fromfile(f'{path_train}/{i}', np.uint8)
    src = cv2.imdecode(src, cv2.IMREAD_GRAYSCALE).astype('float32')
    try:
        imwrite(f'{path_train}_lr2/{i}', cv2.resize(cv2.resize(src, None, fx=0.1, fy=0.1), None, fx=5, fy=5))
    except:
        print(i)
```

```python
for i in os.listdir('crop'):
    src = np.fromfile(f'{path_train}_lr/{i}', np.uint8)
    src = cv2.imdecode(src, cv2.IMREAD_GRAYSCALE).astype('float32')
    src2 = np.fromfile(f'result_image_finetune/{i}', np.uint8)
    src2 = cv2.imdecode(src2, cv2.IMREAD_GRAYSCALE).astype('float32')
    plt.subplot(1, 2, 1)
    plt.imshow(src, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(src2, cmap='gray')
    plt.show()
```
